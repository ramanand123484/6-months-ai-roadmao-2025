# For else:-
   # For|Else in python
   # In other language for else not supportable but in python it is supportable
# e.g:- lets print the number from 1-20 & we dont want print number which is divisible by 5

```python
nums = [12,15,18,21,26,30,40]

for num in nums:
    if num%5==0:
        print(num)
```

```python
nums = [12,14,18,21,25,30,35]

for num in nums:
    if num%5==0:
        print(num)
```

```python
nums = [12,14,18,21,20,25]  # here check 12,14,18,21 this test case is faile after go to 20 this case is satisfy then print 20 and break the loop

for num in nums:
    if num%5==0:
        print(num)
        break
```

```python
nums = [7,14,18,21,23,27] #here is no number which is divisible by 5 we got output as blank

for num in nums:
    if num%5==0:
        print(num)
        break
```

```python
nums = [7,14,18,21,23,27,29] # here there is no number which is divisble by 5 we got output as Number Not Found

for num in nums:
    if num%5==0:
        print(num)
        break
else:
 print('Number Not Found') # every iteration it checking condition.
```
